```python
data ={'name':['aaa','bbb','zzz','sss','qqq','www','dddd'],
          'gender':['male','male','female','male','female','female','male'],
           'class':['CS_101','CS_102','CS_101','CS_102','CS_101','CS_101','CS_102'],
          'chi':[80,63,26,75,78,90,90],
          'eng':[90,98,97,67,58,90,70],
          'math':[20,40,50,60,70,80,49]}
df=pd.DataFrame(data,index=list('abcdefg'),columns=['name','class','gender','chi','eng','math'])
df

```

```python
# split -DataFrameGroupby
g = df.groupby(df['class'])
g=df.groupby('class')
for k,v in g: #k= key ,v=DataFrame
    print(k)
    print(v)
    print('---------------------------')
```
